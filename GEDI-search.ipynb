```python
gedi_product_short_name = "GEDI02_B.002"
start_date = "2019-04-18"
end_date = "2023-03-16" # Adjust based on GEDI mission operational periods, cuz GEDI instrument  moved to ISS on March 17, 2023 * 
lower_left_lon = bbox[0]
lower_left_lat = bbox[1]
upper_right_lon = bbox[2]
upper_right_lat = bbox[3]
print(f"Searching for {gedi_product_short_name} data within "
      f"lon: {lower_left_lon} to {upper_right_lon}, lat: {lower_left_lat} to {upper_right_lat} "
      f"for the period {start_date} to {end_date}...")

results = earthaccess.search_data(
    short_name=gedi_product_short_name,
    bounding_box=(lower_left_lon, lower_left_lat, upper_right_lon, upper_right_lat),
    temporal=(start_date, end_date)
)
```
